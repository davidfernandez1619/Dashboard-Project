```python
import os
import pandas as pd

current_directory = os.getcwd()
print("Directorio actual:", current_directory)
```

```python
PIB = pd.read_csv('/Users/davidfernandez/Desktop/PIB/1.csv', sep=';')

PIB
```

```python
PIB.isnull().sum() # Our dataframe doesn't contain any null values, we are good to go
```

```python
columns_to_drop = ['region_name', 'sub_region_name', 'intermediate_region', 'total_gdp', 'gdp_variation']
# These are the columns that we want to remove, since we're only interested in countries, and their total gdp by million

PIB = PIB.drop(columns_to_drop, axis=1) # We drop the unwanted columns
```

```python
PIB['income_group'].unique() # We want to translate these values to english
```

```python
translation_dict = {
    'Ingreso alto': 'High income',
    'Países de ingreso bajo': 'Low income',
    'Países de ingreso mediano bajo': 'Mid-low income',
    'Ingreso mediano alto': 'Mid-high income',
    'No clasificado': 'Not specified'}

PIB['income_group'] = PIB['income_group'].replace(translation_dict)

# We replace the spanish terms to english terms

```

```python
condition = (PIB['year'] >= 1990) & (PIB['year'] <= 2017)

PIB = PIB[condition]

PIB = PIB.reset_index()

PIB = PIB.drop('index', axis=1)

PIB

# Finally, we filter our dataframe to get the countries between 1990 and 2017, and we reset indexes
```
