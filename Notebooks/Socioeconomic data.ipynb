## Unemployment

```python
paro = pd.read_csv('/Users/davidfernandez/Desktop/Datos socioeconómicos/paro.csv', encoding='ISO-8859-1')

paro # Our data has the same problem as our previous 'Consumer Price Indices' csv
```

```python
new_paro = paro.reset_index()
divided_columns = new_paro['index'].str.split('\t', expand=True)
index = divided_columns[4] 
second_column = new_paro.iloc[:, 1]  
new_column = index + ',' + second_column.astype(str)
new_paro['Indices (%)'] = new_column
```

```python
columns = new_paro['index'].str.split('\t', expand=True)

date = divided_columns[3]

new_paro['index'] = date
```

```python
columns_to_remove = ['Sexo\tComunidades y Ciudades Autónomas\tEdad\tPeriodo\tTotal']

new_paro = new_paro.drop(columns_to_remove, axis=1)
```

```python
new_paro['Year'] = new_paro['index'].str[:4].astype(int)
new_paro['Month'] = new_paro['index'].str[4:]

# We create two new columns with the year and each month

column_order = ['index', 'Year', 'Month', 'Indices (%)']  # We set the new column order

new_paro = new_paro[column_order] # We create the new column order

new_paro.drop(columns=['index'], inplace=True) # And we get rid of our 'index' column

new_paro['Indices (%)'] = new_paro['Indices (%)'].str.replace(',', '.').astype(float) # And we change the values to float type
```

```python
paro = new_paro.groupby('Year')['Indices (%)'].mean().reset_index()

paro.to_csv('/Users/davidfernandez/Desktop/Projects/Dashboard-Project/data/Unemployment.csv', index=False)
```
