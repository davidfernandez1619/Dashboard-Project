## Cleaning our DataFrame

```python
import pandas as pd
```

```python
mental = pd.read_csv('/Users/davidfernandez/Desktop/mental.csv') # This is our first CSV
```

```python
mental
```

```python
print(mental[mental.Year == 'Year'])
''' As we can see here, in our dataframe there is 4 different dataframes. The word 'Year'
is what separates one dataframe from another. The first df is between rows 1 and 6467, the second df is between
rows 6488 and 54275, the third df is between rows 54276 and 102083 and the fourth one is from that row to the last. 
Each df is concatenated to the previous one and, since all columns don't match, there is a lot of null values'''
```

```python
df1 = mental.iloc[:6468].copy() # We take the first 6467 rows and store them in our first dataframe
```

```python
nulls = df1.isnull().sum() # We check if our dataframe contains null values. It does, in the 'Code' column
nulls
```

```python
df1['Entity'].loc[df1['Code'].isnull()].unique()
# We want to check, from the column 'code', which countries have null values. 
# As we can see, the values with null values are not countries, but groups of countries that don't have specific codes
```

```python
df1.dropna(axis=0, inplace = True) # So we will just get rid of the null values
```

```python
column_to_drop = ['index']
df1 = df1.drop(column_to_drop, axis=1) # So we get rid of the column 'index', since it won't be useful for us

new_name = {'Entity': 'Country'}
df1.rename(columns=new_name, inplace=True) # And we change the name of 'Entity' column for 'Country'
```

```python
df1.info() # Here we can see that some numeric columns in our dataframe are 'objetct' type, and we want them to be numeric

df1.iloc[:, 2:] = df1.iloc[:, 2:].apply(pd.to_numeric) # We convert those columns (since the third column) to numeric values
```

```python
df1.info() # Now we can see that the columns that have the prevalence data are 'float' type
```

```python
df_filtered = df1[df1['Country'] == 'Afghanistan']

df_filtered['Year'].unique()

# In this code, we can see that our dataframe shows data from the year 1990 until 2017

```

```python
df1['Country'].unique()
```

```python
df1 # This will be our final dataframe, so we are going to save it as a CSV

df1.to_csv('/Users/davidfernandez/Desktop/Projects/Dashboard-Project/cleaned.csv', index=False)
```
